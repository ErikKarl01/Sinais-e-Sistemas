## Questão 4 ítem d

```python
import numpy as np
import matplotlib.pyplot as plt

dt = 0.01
t = np.arange(-1, 5, dt)
x = np.where((t >= 0) & (t < 2), 1.0, 0.0)
h = np.where(t >= 0, np.exp(-2*t), 0.0)

y = np.convolve(x, h) * dt
t_y = np.arange(2 * t[0], 2 * t[-1] + dt, dt)
t_y = t_y[:len(y)]

plt.figure(figsize=(8, 4))
plt.plot(t, x, label='Entrada x(t)', color='blue', linestyle='dashed', linewidth=2)
plt.plot(t_y, y, label='Saída y(t)', color='red', linewidth=2)
plt.xlim(-1, 5)
plt.title('Questão 4 (d): Entrada e Saída')
plt.xlabel('Tempo (t)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

## Questão 4 ítem e

```python
import numpy as np
import matplotlib.pyplot as plt

dt = 0.01
t = np.arange(-1, 5, dt)
x = np.where((t >= 0) & (t < 2), 1.0, 0.0)
h = np.where(t >= 0, np.exp(t), 0.0)

y = np.convolve(x, h) * dt
t_y = np.arange(2 * t[0], 2 * t[-1] + dt, dt)
t_y = t_y[:len(y)]

plt.figure(figsize=(8, 4))
plt.plot(t, x, label='Entrada x(t)', color='blue', linestyle='dashed', linewidth=2)
plt.plot(t_y, y, label='Saída y(t) (Instável)', color='red', linewidth=2)
plt.xlim(-1, 5)
plt.title('Questão 4 (e): Sistema Instável')
plt.xlabel('Tempo (t)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

## Questão 5 ítem b
