```python
import numpy as np
import matplotlib.pyplot as plt
```

$$x(t) = \begin{cases} 2, & -1.5 \le t < 0 \\ 2e^{-t/2}, & 0 \le t \le 3 \\ 0, & \text{caso contrário} \end{cases}$$

```python
def x_func(t):
    # Condição para a parte constante
    cond1 = (t >= -1.5) & (t < 0)
    # Condição para a parte exponencial
    cond2 = (t >= 0) & (t <= 3)
    
    return np.where(cond1, 2.0, np.where(cond2, 2.0 * np.exp(-t/2), 0.0))

# Criando o vetor de tempo para a base
t_base = np.linspace(-8, 10, 2000)
```

```python
# ---------------------------------------------------------
# (b) e (c) Plotagem e Transformações
# ---------------------------------------------------------
```

```python
# Definindo as transformações
sinais = [
    (t_base, x_func(t_base), "Original: $x(t)$"),
    (t_base, x_func(-t_base), "(i) Reversão: $x(-t)$"),
    (t_base, x_func(t_base + 4), "(ii) Deslocamento: $x(t+4)$"),
    (t_base, x_func(t_base - 2), "(iii) Deslocamento: $x(t-2)$"),
    (t_base, x_func(2 * t_base), "(iv) Compressão: $x(2t)$"),
    (t_base, x_func(t_base / 2), "(v) Expansão: $x(t/2)$")
]

# Configuração da figura
fig, axes = plt.subplots(3, 2, figsize=(13, 10), constrained_layout=True)
axes = axes.flatten()

for i, (t_plot, x_plot, titulo) in enumerate(sinais):
    axes[i].plot(t_plot, x_plot, color='blue', lw=2)
    axes[i].set_title(titulo, fontsize=12, fontweight='bold')
    axes[i].axhline(0, color='black', lw=1)
    axes[i].axvline(0, color='black', lw=1)
    axes[i].grid(True, linestyle='--', alpha=0.6)
    axes[i].set_xlim(-6, 8)
    axes[i].set_ylim(-0.2, 2.5)

plt.show()
```

### ANÁLISE DAS OPERAÇÕES NA VARIÁVEL INDEPENDENTE:
(i)   $x(-t)$:   Reversão Temporal. Espelha o sinal no eixo $y$.  
(ii)  $x(t+4)$:  Deslocamento (Avanço). Move o sinal 4 unidades para a ESQUERDA.  
(iii) $x(t-2)$:  Deslocamento (Atraso). Move o sinal 2 unidades para a DIREITA.  
(iv)  $x(2t)$:   Compressão Temporal. O sinal ocorre 2x mais rápido (encolhe no eixo $t$).  
(v)   $x(t/2)$:  Expansão Temporal. O sinal ocorre na metade da velocidade (estica no eixo $t$).
