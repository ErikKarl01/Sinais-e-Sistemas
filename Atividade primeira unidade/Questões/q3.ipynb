```python
import numpy as np
import matplotlib.pyplot as plt

t_start = 0
t_end = 0.1
t = np.linspace(t_start, t_end, 2000) 

A = 1.5
fm = 50  # Hz
fc = 500 # Hz

# Geração dos sinais
x_t = np.cos(2 * np.pi * fm * t)            # Mensagem
y_t = (A + x_t) * np.cos(2 * np.pi * fc * t) # Sinal Modulado
```

```python
plt.figure(figsize=(12, 6))

# Plot do sinal modulado y(t)
plt.plot(t, y_t, color='blue', label='Saída Modulada $y(t)$')

# Plot da envoltória (A + x(t)) para conferência visual
plt.plot(t, A + x_t, 'r--', alpha=0.6, label='Envoltória Superior')
plt.plot(t, -(A + x_t), 'r--', alpha=0.6, label='Envoltória Inferior')

plt.title('Questão 3: Sinal Modulado em Amplitude (AM)')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend(loc='upper right')
plt.xlim(0, 0.1)

plt.show()
```

* Frequência: Note que existem 5 ciclos completos da mensagem $x(t)$ no intervalo de $0.1$s (pois $f_m = 50$ Hz, logo o período é $0.02$ s).  
* Modulação: A amplitude da portadora (os picos azuis) segue exatamente o formato da onda senoidal da mensagem, deslocada pelo nível DC ($A = 1.5$).  
* Índice de Modulação: Como o valor máximo de $|x(t)|$ é $1$ e $A = 1.5$, não ocorre sobremodulação, pois $A > |x(t)|_{max}$, mantendo a envoltória sempre acima de zero.
