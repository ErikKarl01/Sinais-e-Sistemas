```python
import numpy as np
import matplotlib.pyplot as plt

# Definição do intervalo de tempo
t = np.linspace(0, 3, 1000)

# Definição do sinal x(t)
# u(t-1) - u(t-2) é 1 entre 1 e 2, e 0 fora disso
x = (t**2 - 1) * ((t >= 1) & (t <= 2))

# Plotagem
plt.figure(figsize=(8, 4))
plt.plot(t, x, label=r'$x(t) = (t^2 - 1)[u(t-1) - u(t-2)]$')
plt.fill_between(t, x, alpha=0.2)
plt.title('Sinal de Tempo Contínuo $x(t)$')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import sympy as sp

t_sym = sp.symbols('t')
x_sym = t_sym**2 - 1

energy_analytic = sp.integrate(x_sym**2, (t_sym, 1, 2))

print(f"Energia Analítica (Exata): {energy_analytic}")
print(f"Energia Analítica (Decimal): {float(energy_analytic):.6f}")
```

```python
energy_numeric = np.trapezoid(x**2, t)

print(f"Energia Numérica: {energy_numeric:.6f}")

# Comparação
erro = abs(float(energy_analytic) - energy_numeric)
print(f"Diferença entre métodos: {erro:.2e}")
```
